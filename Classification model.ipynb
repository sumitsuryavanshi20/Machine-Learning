Decision Tree Classification

```python
import pandas as pd
```

```python
Dataset = pd.read_csv("tennis.csv")
Dataset
```

```python
from sklearn.preprocessing import LabelEncoder
```

```python
# Create a LabelEncoder object
le = LabelEncoder()
```

```python
# Apply LabelEncoder to each categorical column
Dataset['outlook'] = le.fit_transform(Dataset['outlook'])
Dataset['temp'] = le.fit_transform(Dataset['temp'])
Dataset['humidity'] = le.fit_transform(Dataset['humidity'])
Dataset['windy'] = le.fit_transform(Dataset['windy'])
Dataset['play'] = le.fit_transform(Dataset['play'])
```

```python
Dataset
```

```python
features_clos=['outlook','humidity','windy']
x=Dataset[features_clos]
y=Dataset.play
```

```python
x
```

```python
y
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train, x_test, y_train, y_test=train_test_split(x,y,test_size=0.2)
```

```python
from sklearn.tree import DecisionTreeClassifier
```

```python
classifier= DecisionTreeClassifier(criterion='gini')
```

```python
classifier.fit(x_train,y_train)
```

```python
classifier.predict(x_test)
```

```python
x_test  #sunny-2, overcast-0, rainy-1
```

```python
classifier.score(x_test, y_test)
```

```python
from sklearn import tree
tree.plot_tree(classifier)
```

Random forest Algorithm

```python
from sklearn.datasets import load_iris
import pandas as pd
```

```python
# Load the iris dataset
iris = load_iris()
```

```python
# Create the DataFrame
dataset1 = pd.DataFrame({
    'sepal length': iris.data[:, 0],
    'sepal width': iris.data[:, 1],
    'petal length': iris.data[:, 2],
    'petal width': iris.data[:, 3],
    'species': iris.target})
```

```python
# Show the DataFrame
print(dataset1.head())
```

```python
x=dataset1[['sepal length', 'sepal width', 'petal length', 'petal width']]
y=dataset1['species']#labels
```

```python
x
```

```python
y
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train, x_test, y_train, y_test=train_test_split(x,y,test_size=0.3)
```

```python
len(x_train)
```

```python
len(x_test)
```

```python
from sklearn.ensemble import RandomForestClassifier
```

```python
clf=RandomForestClassifier(n_estimators=100, criterion='gini')
```

```python
clf.fit(x_train,y_train)
```

```python
clf.predict(x_test)
```

```python
x_test  #0 stosa, 1 versiccala, 2 vaginica
```

```python
clf.score(x_test, y_test)
```

```python
clf.predict([[3,5,4,2]])
```

```python
clf.predict([[3,3,2,2,]])
```

```python
feature_imp=pd.Series(clf.feature_importances_,index=iris.feature_names).sort_values(ascending=False)
```

```python
feature_imp
```
